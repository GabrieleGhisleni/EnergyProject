```python
# import pandas as pd
from datetime import timedelta
df = pd.read_csv("merged.csv")
```

```python
df["date_x"]=pd.to_datetime(df["date_x"],format= "%d/%m/%Y %H:%M:%S %p")
df.shape
```

```python
x=pd.to_datetime(df["cross_join"], format="%d/%m/%Y %H:%M %p")
df["cross_join"]=x.dt.strftime("%Y-%m-%d %H:%M:%S")
df = df[df["cross_join"].str.endswith("00:00")]
df["cross_join"]=pd.to_datetime(df["cross_join"], format="%Y-%m-%d %H:%M:%S")
```

```python
def cheat_days(x):
    day = timedelta(days=5)
    return (x-day)
df["cross_join"] = df["cross_join"].apply(cheat_days)
df["cross_join"]=df["cross_join"].dt.strftime("%Y-%m-%d %H:%M:%S")
```

```python
def clean_rain(x):
    if type(x) == str and '{' in x:
        index = x.find(":")
        return x[index+1:-1].strip()
    else:
        return x
    
df["rain_1h"] = df["rain_1h"].apply(clean_rain)
```
